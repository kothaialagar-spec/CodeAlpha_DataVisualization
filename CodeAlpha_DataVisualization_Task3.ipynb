```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

iris = load_iris()

df = pd.DataFrame(
    iris.data,
    columns=iris.feature_names
)

df.head()
```

```python
df.info()
```

```python
df['sepal length (cm)'].hist()

plt.title('Sepal Length Distribution')

plt.xlabel('Sepal Length')

plt.ylabel('Frequency')

plt.show()
```

Observation:
The sepal length values are distributed across a moderate range, with most flowers concentrated around the middle values.

```python
df.mean().plot(kind='bar')

plt.title('Average Feature Values')

plt.ylabel('Average Value')

plt.show()
```

Observation:
The average values of flower features vary, indicating differences among the measured characteristics.

```python
df.iloc[:20].plot()

plt.title('Feature Trends')

plt.show()
```

Observation:
The line chart shows how different flower measurements vary across the first 20 samples.

```python
sizes = [50, 50, 50]
labels = ['Setosa', 'Versicolor', 'Virginica']

plt.pie(sizes, labels=labels, autopct='%1.1f%%')

plt.title('Iris Species Distribution')

plt.show()
```

Observation:
The dataset contains an equal number of samples from all three Iris species.

```python
import seaborn as sns
```

```python
plt.figure(figsize=(8,6))

sns.heatmap(df.corr(), annot=True)

plt.title('Correlation Heatmap')

plt.show()
```

Observation:
The heatmap reveals relationships between different flower measurements and helps identify highly correlated features.

Findings:

1. The Iris dataset contains measurements from three flower species.
2. Sepal length values are moderately distributed.
3. Feature averages differ across measurements.
4. Species are equally represented in the dataset.
5. Correlation analysis reveals relationships among features.
6. Data visualization makes patterns easier to identify and interpret.
